```python
#### Declarations of process functions for Optical Propertises

### Necessary packages invoking
import numpy as np
import xml.etree.ElementTree as ET

### Physical constants
hbar = 4.135667662e-15
c_vacuum = 2.99792458e8

### Extract optical propertises from vasprun.xml
def extract_opt_vectorized(file_path):
    ## Analysis vasprun.xml file
    tree = ET.parse(file_path)
    root = tree.getroot()

    ## Extract NEDOS
    nedos_element = root.find(".//i[@name='NEDOS']")
    nedos_number = int(nedos_element.text.strip())

    ## Loop variables
    # Define prefixes: e_ for Density-Density, c_ for Current-Current
    prefixes = ["e_", "c_"]
    # Initialize a dictionary to store dynamic variables
    data = {}

    ## Extract imaginary part of Density-Density and Current-Current
    path_imag = ".//dielectricfunction/imag/array/set"
    imag_elements = root.findall(path_imag)

    for loop_index, imag_element in enumerate(imag_elements[:2]):
        # Select prefix based on the loop index
        prefix = prefixes[loop_index]
        # Initialize columns
        data[prefix + 'energy_imag_col'] = np.empty(0)
        data[prefix + 'xx_imag_col'] = np.empty(0)
        data[prefix + 'yy_imag_col'] = np.empty(0)
        data[prefix + 'zz_imag_col'] = np.empty(0)
        data[prefix + 'xy_imag_col'] = np.empty(0)
        data[prefix + 'yz_imag_col'] = np.empty(0)
        data[prefix + 'zx_imag_col'] = np.empty(0)
        for imag_index in imag_element.findall("r"):
            imag_values = list(map(float, imag_index.text.split()))
            # Append data to columns
            data[prefix + 'energy_imag_col'] = np.append(data[prefix + 'energy_imag_col'], imag_values[0])
            data[prefix + 'xx_imag_col'] = np.append(data[prefix + 'xx_imag_col'], imag_values[1])
            data[prefix + 'yy_imag_col'] = np.append(data[prefix + 'yy_imag_col'], imag_values[2])
            data[prefix + 'zz_imag_col'] = np.append(data[prefix + 'zz_imag_col'], imag_values[3])
            data[prefix + 'xy_imag_col'] = np.append(data[prefix + 'xy_imag_col'], imag_values[4])
            data[prefix + 'yz_imag_col'] = np.append(data[prefix + 'yz_imag_col'], imag_values[5])
            data[prefix + 'zx_imag_col'] = np.append(data[prefix + 'zx_imag_col'], imag_values[6])

    return (nedos_number, data['e_energy_imag_col'])

## File path (you should update this to the correct file path)
opt_file_path = "7_Optical_properties_gamma/vasprun.xml"

result_testing = extract_opt_vectorized(opt_file_path)

print(result_testing)
```
