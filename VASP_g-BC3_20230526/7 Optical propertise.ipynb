```python
#### Declarations of process functions for Optical Propertises

### Necessary packages invoking
import numpy as np
import xml.etree.ElementTree as ET

### Physical constants
hbar = 4.135667662e-15
c_vacuum = 2.99792458e8

### Extract optical propertises from vasprun.xml
def extract_opt_vectorized(file_path):
    ## Analysis vasprun.xml file
    tree = ET.parse(file_path)
    root = tree.getroot()

    ## Extract NEDOS
    nedos_element = root.find(".//i[@name='NEDOS']")
    nedos_number = int(nedos_element.text.strip())

    ## Loop variables
    # Define prefixes: e_ for Density-Density, c_ for Current-Current
    prefixes = ["e_", "c_"]

    ## Extract imaginary part of Density-Density and Current-Current
    path_imag = ".//dielectricfunction/imag/array/set"
    imag_elements = root.findall(path_imag)

    for loop_index, imag_element in enumerate(imag_elements[:2]):
        # Select prefix based on the loop index
        prefix = prefixes[loop_index]
        # Initialize columns
        exec(f"{prefix}energy_imag_col = np.empty(0)")
        exec(f"{prefix}xx_imag_col = np.empty(0)")
        exec(f"{prefix}yy_imag_col = np.empty(0)")
        exec(f"{prefix}zz_imag_col = np.empty(0)")
        exec(f"{prefix}xy_imag_col = np.empty(0)")
        exec(f"{prefix}yz_imag_col = np.empty(0)")
        exec(f"{prefix}zx_imag_col = np.empty(0)")
        for imag_index in imag_element.findall("r"):


    # Columns initialization
    for imag_element in root.findall(path_imag):
        for imag_index in imag_element.findall("r"):
            imag_values = list(map(float, imag_index.text.split()))
            # Columns of energy
            e_energy_imag_col = np.append(e_energy_imag_col, imag_values[0])
            e_xx_imag_col = np.append(e_xx_imag_col, imag_values[1])
            e_yy_imag_col = np.append(e_yy_imag_col, imag_values[2])
            e_zz_imag_col = np.append(e_zz_imag_col, imag_values[3])
            e_xy_imag_col = np.append(e_xy_imag_col, imag_values[4])
            e_yz_imag_col = np.append(e_yz_imag_col, imag_values[5])
            e_zx_imag_col = np.append(e_zx_imag_col, imag_values[6])
        
    ## Extract real part of Density-Density
    # Columns initialization
    e_energy_real_col = np.empty(0)
    e_xx_real_col = np.empty(0)
    e_yy_real_col = np.empty(0)
    e_zz_real_col = np.empty(0)
    e_xy_real_col = np.empty(0)
    e_yz_real_col = np.empty(0)
    e_zx_real_col = np.empty(0)
    path_real = ".//dielectricfunction/real/array/set"
    real_element = root.find(path_real)
    for real_index in real_element.findall("r"):
        real_values = list(map(float, real_index.text.split()))
        # Columns of energy
        e_energy_real_col = np.append(e_energy_real_col, real_values[0])
        e_xx_real_col = np.append(e_xx_real_col, real_values[1])
        e_yy_real_col = np.append(e_yy_real_col, real_values[2])
        e_zz_real_col = np.append(e_zz_real_col, real_values[3])
        e_xy_real_col = np.append(e_xy_real_col, real_values[4])
        e_yz_real_col = np.append(e_yz_real_col, real_values[5])
        e_zx_real_col = np.append(e_zx_real_col, real_values[6])

    return (nedos_number, e_xx_imag_col)

## File path (you should update this to the correct file path)
opt_file_path = "7_Optical_properties_gamma/vasprun.xml"

result_testing = extract_opt_vectorized(opt_file_path)[1]

print(result_testing)
print(len(result_testing))
```
