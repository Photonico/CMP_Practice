# Data analysis script for VASP

## Band data

```python
# Band data

def extract_band_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    bands = []
    k_points = []
    for i, line in enumerate(lines):
        if 'k-point' in line:
            k_point = list(map(float, line.split()[3:6]))
            k_points.append(k_point)
            band_energies = []
            j = i + 1
            while 'band' in lines[j]:
                band_energy = float(lines[j].split()[-2])
                band_energies.append(band_energy)
                j += 1
            bands.append(band_energies)

    return k_points, bands

vasp_out_file = "../vasp.out"
k_points, bands = extract_band_data(vasp_out_file)

import matplotlib.pyplot as plt

def plot_bands(k_points, bands):
    k_point_distances = [0]
    for i in range(1, len(k_points)):
        distance = ((k_points[i][0] - k_points[i-1][0])**2 +
                    (k_points[i][1] - k_points[i-1][1])**2 +
                    (k_points[i][2] - k_points[i-1][2])**2)**0.5
        k_point_distances.append(k_point_distances[-1] + distance)

    for band in zip(*bands):
        plt.plot(k_point_distances, band, marker='o')

    plt.xlabel("k-points")
    plt.ylabel("Energy (eV)")
    plt.title("Band structure")
    plt.grid()
    plt.show()

plot_bands(k_points, bands)
```

## Energy data

```python
# Energy data

def extract_energies(filename):
    energies = []
    with open(filename, 'r') as file:
        for line in file:
            if 'energy  without entropy=' in line:
                energy = float(line.split()[-2])
                energies.append(energy)
    return energies

vasp_out_file = "../vasp.out"
energies = extract_energies(vasp_out_file)

import matplotlib.pyplot as plt

def plot_energies(energies):
    plt.plot(energies, marker='o')
    plt.xlabel("Optimization step")
    plt.ylabel("Energy (eV)")
    plt.title("Energy vs Optimization step")
    plt.grid()
    plt.show()

plot_energies(energies)
```
